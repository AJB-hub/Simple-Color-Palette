```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# n = number of subdivisions of the color wheel
subdiv = 20 #int(input("How many subdivisions?: "))
colors = 0
wheel = range(2,subdiv+1)
b = 0
c = 0
```

```python
def cos(color):
    color = np.cos(np.pi*(colors/subdiv))
    return color
```

```python
color_df = pd.DataFrame(columns=['color','red', 'green', 'blue'])
color_df = color_df.append({'color':1,'red':1, 'green':b, 'blue':c},ignore_index=True)
```

```python
for colors in wheel:
    color_df = color_df.append({
                            'color': colors, 
                            'red': a,
                            'green': b,
                            'blue': c
                                
        
    },ignore_index=True)
    if color_df["color"] >= subdiv/3:
        b = np.cos(np.pi*(colors/subdiv))
        a = 0
        c = 0
        
```
